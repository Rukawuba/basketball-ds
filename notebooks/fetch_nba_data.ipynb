```python
from nba_api.stats.endpoints import LeagueGameLog
import pandas as pd
```

```python
log = LeagueGameLog(season='2024-25', season_type_all_star='Regular Season')
df = log.get_data_frames()[0]
df.head()
```

```python
df.to_csv("../data/league_gamelog_2024.csv", index=False)
```

```python
from nba_api.stats.endpoints import CommonAllPlayers
players = CommonAllPlayers(is_only_current_season=1).get_data_frames()[0]
players.to_csv("../data/players_2024.csv", index=False)
players.head()
```

```python
from nba_api.stats.endpoints import TeamDetails
import pandas as pd

team_ids = players["TEAM_ID"].dropna().unique()

team_data = []
for tid in team_ids:
    detail = TeamDetails(team_id=int(tid)).get_data_frames()[0]
    team_data.append(detail)

teams_df = pd.concat(team_data, ignore_index=True)
teams_df.to_csv("../data/teams_2024.csv", index=False)
teams_df.head()
```

```python
from nba_api.stats.endpoints import LeagueGameLog
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "application/json, text/plain, */*",
    "Accept-Language": "en-US,en;q=0.9",
    "Origin": "https://www.nba.com",
    "Referer": "https://www.nba.com/",
    "Connection": "keep-alive",
}

def fetch_games(season="2023-24"):
    for attempt in range(5):
        try:
            print(f"Attempt {attempt+1}…")
            log = LeagueGameLog(
                season=season,
                season_type_all_star="Regular Season",
                timeout=60,
                headers=headers
            )
            return log.get_data_frames()[0]
        except Exception as e:
            print("Failed:", e)
            time.sleep(4)

games = fetch_games()
games.head()
```
