```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = {
    "player": ["Booker", "KD", "Allen", "Nurkic", "Beal"],
    "pts": [28, 27, 15, 12, 20],
    "ast": [6, 5, 3, 4, 7],
    "min": [36, 35, 32, 30, 34]
}

df = pd.DataFrame(data)
df

```

```python
df["pts_per_min"] = df["pts"] / df["min"]
df
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df, x="player", y="pts")
plt.title("Points per Player")
plt.tight_layout()
plt.show()
```
